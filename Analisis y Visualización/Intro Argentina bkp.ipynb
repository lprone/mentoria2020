```python
df_datasource = pd.read_csv('https://raw.githubusercontent.com/lprone/mentoria2020/master/dataset/sistemasMapache/Covid19arData%20-%20covid19argentina_datasource.csv')
```

```python
del df_datasource['osm_admin_level_2']
del df_datasource['SUMtot_recuperados']
del df_datasource['SUMtot_test_negativos']
del df_datasource['SUMtot_test']
```

```python
df_datasource.head()
```

```python
df_datasource[df_datasource['covid19argentina_admin_level_4'] == 'no-data']
```

```python
df_datasource.drop(df_datasource[df_datasource.covid19argentina_admin_level_4 == 'no-data'].index, inplace=True)
df_datasource.drop(df_datasource[df_datasource.covid19argentina_admin_level_4.isna()].index, inplace=True)
```

```python
df_datasource.fecha = pd.to_datetime(df_datasource.fecha.astype(str), format='%d/%m/%Y')
```

```python
df_datasource.head()
```

```python
df_datasource.rename(columns={
    'fecha':'index',
    'covid19argentina_admin_level_4':'Province/State'
}, inplace=True)
```

```python
df_datasource.head()
```

```python
df_datasource.sort_values(by=['Province/State', 'index'], ascending=True)
```

```python
#df_datasource = df_datasource.groupby(['Province/State', 'index']).sum().groupby(level=0).cumsum().reset_index()
```

```python
df_deaths = df_datasource[['index','Province/State','SUMnue_fallecidos_diff']]
df_confirmed = df_datasource[['index','Province/State','SUMnue_casosconf_diff']]
```

```python
df_deaths.head(20)
```

```python
df_confirmed.head()
```

```python
df_confirmed.head()
```

```python
df_confirmed['index'] = df_confirmed["index"].dt.strftime('%m/%d/%y')
df_deaths['index'] = df_deaths["index"].dt.strftime('%m/%d/%y')
```

```python
df_deaths.head()
```

```python
df_confirmed_pivot = df_confirmed.pivot(index='Province/State', columns='index', values='SUMnue_casosconf_diff').fillna(0)
df_deaths_pivot = df_deaths.pivot(index='Province/State', columns='index', values='SUMnue_fallecidos_diff').fillna(0)
```

```python
df_confirmed_pivot.head()
```

```python
df_confirmed_pivot.insert(0, 'Country/Region', 'Argentina')
df_deaths_pivot.insert(0, 'Country/Region', 'Argentina')
```

```python
df_confirmed_pivot.head()
```

```python
df_confirmed_pivot = df_confirmed_pivot.reset_index()
df_deaths_pivot = df_deaths_pivot.reset_index()
```

```python
df_confirmed_pivot
```

```python
for r in range(0,len(df_confirmed_pivot)):
    for c in range(3,len(df_confirmed_pivot.columns)):
        df_confirmed_pivot.iloc[r,c] = df_confirmed_pivot.iloc[r,c - 1] + df_confirmed_pivot.iloc[r,c]
```

```python
df_confirmed_pivot
```

```python
df_deaths_pivot
```

```python
for r in range(0,len(df_deaths_pivot)):
    for c in range(3,len(df_deaths_pivot.columns)):
        df_deaths_pivot.iloc[r,c] = df_deaths_pivot.iloc[r,c - 1] + df_deaths_pivot.iloc[r,c]
```
