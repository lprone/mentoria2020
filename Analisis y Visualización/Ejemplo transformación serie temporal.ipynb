```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import pandas as pd
```

```python
data_folder = '../dataset/johnsHopkins/'

confirmed_filename = 'time_series_covid19_confirmed_global.csv'
deaths_filename = 'time_series_covid19_deaths_global.csv'

confirmed_path = data_folder + confirmed_filename
deaths_path = data_folder + deaths_filename
```

```python
df_confirmed = pd.read_csv(confirmed_path)
df_deaths = pd.read_csv(deaths_path)
```

```python
df_confirmed.head()
```

```python
df_confirmed[df_confirmed['Country/Region'] == 'Argentina']
```

```python
df_deaths[df_deaths['Country/Region'] == 'Argentina']
```

```python
df_confirmed_time_series = df_confirmed
```

```python
df_confirmed_time_series.head()
```

```python
del df_confirmed_time_series['Province/State']
del df_confirmed_time_series['Lat']
del df_confirmed_time_series['Long']
```

```python
df_confirmed_time_series = df_confirmed_time_series\
                .groupby('Country/Region', as_index=False)\
                .sum()\
                .sort_values("Country/Region")
```

```python
df_confirmed_time_series.head()
```

```python
df_confirmed_time_series.rename(columns={'Country/Region':'Date'}, inplace=True)
```

```python
df_confirmed_time_series = df_confirmed_time_series.T
```

```python
df_confirmed_time_series.head()
```

```python
df_confirmed_time_series.columns = df_confirmed_time_series.iloc[0]
```

```python
df_confirmed_time_series.head()
```

```python
df_confirmed_time_series = df_confirmed_time_series.iloc[1:]
```

```python
df_confirmed_time_series.head()
```

```python
df_confirmed_time_series.index = pd.to_datetime(df_confirmed_time_series.index.astype(str), format='%m/%d/%y')
```
